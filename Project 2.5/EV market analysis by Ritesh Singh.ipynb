```python
#predicting next 3 years sale using linear regression

from sklearn.linear_model import LinearRegression

X = np.array([year for year in range(2015, 2025)]).reshape(-1, 1)  # Years as independent variable
predictions = {}

for category in ev_sales_by_vehicle_category.columns:
    y = ev_sales_by_vehicle_category[category].values.reshape(-1, 1)  # Sales data
    
    model = LinearRegression()
    model.fit(X, y)
    
    future_years = np.array([2025, 2026, 2027]).reshape(-1, 1)
    future_sales = model.predict(future_years)
    
    predictions[category] = future_sales.flatten()

for category, sales in predictions.items():
    print(f"Predicted sales for {category} in 2025-2027: {sales}")
```

```python
# Predicted years
years = np.array([2025, 2026, 2027])

# Predicted sales for each vehicle category
sales_data = {
    '2W': [711902.13, 799894.38, 887886.62],
    '3W': [476565.13, 528044.92, 579524.71],
    'LMV': [64221.67, 72145.82, 80069.98],
    'MMV': [182.73, 201.34, 219.95]
}

plt.figure(figsize=(10, 6))

for category, sales in sales_data.items():
    plt.plot(years, sales, marker='o', label=category)

plt.xlabel('Year')
plt.ylabel('Predicted Sales')
plt.title('Predicted EV Sales Trends (2025-2027)')
plt.xticks(years)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Convert 'Date' column to datetime
evcat["Date"] = pd.to_datetime(evcat["Date"], format="%d/%m/%y", errors="coerce")
evcat
```

```python
# Extract year
evcat["Year"] = evcat["Date"].dt.year
evcat
```

```python
# Aggregate sales by year for each category
yearly_sales = evcat.groupby("Year").sum()  # Summing over all months
yearly_sales
```

```python
# Transpose the data so vehicle categories are rows, years are columns
transposed_sales = yearly_sales.T
transposed_sales
```

```python
# Standardize the data for clustering

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(transposed_sales)
print(scaled_data[:5])  # Display first 5 rows
```

```python
# Hierarchical Clustering
from scipy.cluster.hierarchy import linkage, dendrogram

linked = linkage(scaled_data, method='ward')

# Plot dendrogram to visualize clusters
plt.figure(figsize=(10, 5))
dendrogram(linked, labels=transposed_sales.index, leaf_rotation=90)
plt.title("Vehicle Category Clustering")
plt.xlabel("Vehicle Categories")
plt.ylabel("Distance")
plt.show()
```

# K-means clustering

```python
from sklearn.cluster import KMeans
```

```python
# Find optimal clusters using the Elbow Method
inertia = []
cluster_range = range(1, 11)  # Test for up to 10 clusters

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
```

```python
# Plot the elbow graph
plt.figure(figsize=(8, 5))
plt.plot(cluster_range, inertia, marker='o', linestyle='--', label='Inertia')

# Add vertical line at the elbow point (around k=3)
plt.axvline(x=3, color='r', linestyle='--', label='Elbow Point')

plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=3, random_state=42)  # Choose the K 
labels = kmeans.fit_predict(scaled_data)

# Compute the Silhouette Score
score = silhouette_score(scaled_data, labels)
print(f"Silhouette Score for 3 Clusters: {score:.4f}")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(scaled_data)

# Compute the Silhouette Score
score = silhouette_score(scaled_data, labels)
print(f"Silhouette Score for 4 Clusters: {score:.4f}")

#observation: silhouette score for k=3 is better
```

```python
# Apply K-Means with optimal k = 3 cluster number
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
transposed_sales['Cluster'] = kmeans.fit_predict(scaled_data)
transposed_sales
```

```python
# Apply PCA to reduce dimensions for visualization
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
pca_data
```

```python
# Create DataFrame to store PCA results and cluster labels
clustered_df = pd.DataFrame(pca_data, columns=["PCA1", "PCA2"])
clustered_df["Cluster"] = labels
clustered_df["Vehicle_Category"] = transposed_sales.index  # Assign vehicle category names
clustered_df

#observation:
'''
- Two Wheeler(NT) (PCA1 = 13.96, PCA2 = -2.70) :
    This category has very high PCA1, 
    meaning it dominates the primary trend. 
    It likely has strong, steady growth over time.

- Three Wheeler(T) (PCA1 = 3.43, PCA2 = 9.67) : 
    High PCA2 suggests it follows a different trajectory, 
    potentially spiking in certain years rather than following a steady trend.

'''
```

# PCA explanation:

- Instead of analyzing every individual year, PCA groups trends into two major dimensions that capture the most variation.
- The vehicle categories are still the same, but now we can cluster them based on their behavior over time rather than raw sales numbers.
- This reduction helps segment vehicles into distinct market patterns, which is useful for clustering.

```python
from scipy.spatial import ConvexHull

# Define a colormap for distinct clusters
palette = {0: "red", 1: "blue", 2: "green"}
markers = {0: "o", 1: "s", 2: "D"}  # Circle, Square, Diamond

# Scatter plot with cluster distinction
plt.figure(figsize=(10, 6))

# Loop through each cluster separately for better visibility
for cluster in clustered_df["Cluster"].unique():
    cluster_data = clustered_df[clustered_df["Cluster"] == cluster]
    
    # Scatter plot for each cluster with distinct markers
    sns.scatterplot(x="PCA1", y="PCA2", data=cluster_data, color=palette[cluster], 
                    marker=markers[cluster], s=150, edgecolor="black", label=f"Cluster {cluster}")


    # Draw convex hulls only if there are at least 3 points
    points = cluster_data[["PCA1", "PCA2"]].values
    if len(points) >= 3:
        hull = ConvexHull(points)
        plt.plot(points[hull.vertices, 0], points[hull.vertices, 1], linestyle="-", linewidth=2, color=palette[cluster], alpha=0.8)


    # Annotate vehicle categories with better placement
    for i in range(len(cluster_data)):
        plt.text(cluster_data["PCA1"].iloc[i], cluster_data["PCA2"].iloc[i], cluster_data["Vehicle_Category"].iloc[i], 
                 fontsize=10, weight="bold", ha="right", va="center")


plt.title("Enhanced Vehicle Category Clustering (PCA Reduced)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster")
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Ensure the sales_series index is in datetime format
sales_series.index = pd.to_datetime(sales_series.index)

# Fit the ARIMA model (you can adjust the order as needed)
model = ARIMA(sales_series, order=(5, 1, 2))
model_fit = model.fit()

# Forecast the next 5 years (or periods)
forecast_steps = 5
forecast = model_fit.forecast(steps=forecast_steps)

# Generate future dates with the same frequency (yearly)
future_dates = pd.date_range(start=sales_series.index[-1] + pd.DateOffset(years=1),
                             periods=forecast_steps, freq='Y')

# Create the plot: using scatter plots removes the connecting lines by default
plt.figure(figsize=(10, 6))

# Historical sales plotted with blue circles
plt.scatter(sales_series.index, sales_series, label="Historical Sales", color="blue", marker="o")

# Forecasted sales plotted with red crosses
plt.scatter(future_dates, forecast, label="Forecasted Sales", color="red", marker="x")

# Annotate historical data points with their sales values
for date, value in zip(sales_series.index, sales_series):
    plt.text(date, value, f'{value}', fontsize=8, color="blue",
             verticalalignment="bottom", horizontalalignment="right")

# Annotate forecasted data points with their predicted sales values (rounded for clarity)
for date, value in zip(future_dates, forecast):
    plt.text(date, value, f'{value:.0f}', fontsize=8, color="red",
             verticalalignment="bottom", horizontalalignment="right")

plt.title("EV Sales Forecast for Two Wheeler (NT)")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
import matplotlib.dates as mdates
import pmdarima as pm

# Ensure TW_sales index is datetime and sorted
TW_sales.index = pd.to_datetime(TW_sales.index)
TW_sales = TW_sales.sort_index()
TW_sales
```

```python
# STEP 2: Model Tuning Using Auto ARIMA (SARIMA for monthly data)
model = pm.auto_arima(
    TW_sales,
    seasonal=True,      # Enable seasonal model
    m=12,               # Assume 12-month seasonality
    trace=True,         # Print model selection details
    error_action='ignore',
    suppress_warnings=True,
    stepwise=True       # Use stepwise algorithm for faster convergence
)

# Print the best-fit model summary
print(model.summary())
```

```python
# STEP 3: Forecasting the Next 18 Months
forecast_periods = 18  # next 18 months
forecast = model.predict(n_periods=forecast_periods)

# Generate future dates starting one month after the last observed date:
future_dates = pd.date_range(
    start=pd.Timestamp("2024-08-01"),
    periods=forecast_periods,
    freq='M'
)

# STEP 4: Plotting ONLY the Forecasted Values (Scatter Plot)
plt.figure(figsize=(10, 6))

# Plot forecasted sales as red scatter markers
plt.scatter(future_dates, forecast, color='red', marker='x', s=100, label='Forecasted Sales')

# Format the x-axis to show month and year (e.g., "Sep 2024")
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=90)

# Annotate forecasted points with their values for clarity
for date, value in zip(future_dates, forecast):
    plt.text(date, value, f'{value:.0f}', fontsize=9, color="red",
             verticalalignment="bottom", horizontalalignment="right")

plt.title("18-Month Forecast for TWO WHEELER (NT)")
plt.xlabel("Month")
plt.ylabel("Forecasted Sales")
plt.legend()
plt.tight_layout()  # Adjust layout for better readability
plt.show()
```
